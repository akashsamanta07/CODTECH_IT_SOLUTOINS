```python
import pandas as pd
```

```python
dataset=pd.read_csv("dataset.csv")
dataset.head(5)
```

```python
dataset.shape
```

```python
dataset = dataset[dataset["Built Year"] >= 1970]
```

```python
dataset.shape
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
dataset=dataset[["number of bedrooms","number of bathrooms","living_area_renov","lot_area_renov","number of floors","grade of the house","Built Year","Number of schools nearby","Distance from the airport","Price"]]
dataset.head(5)
```

```python
dataset.info()
```

```python
def roundfig(item):
    return round(item)
```

```python
dataset["number of bathrooms"]=dataset["number of bathrooms"].apply(roundfig)
```

```python
dataset["number of floors"]=dataset["number of floors"].apply(roundfig)
```

```python
dataset
```

```python
dataset.describe()
```

```python
dataset.columns
```

```python
input=dataset.drop(columns=["Price"])
output=dataset["Price"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
```

```python
model=RandomForestRegressor()
```

```python
x_train, x_test, y_train, y_test=train_test_split(input,output,test_size=0.2,random_state=42)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
r2=r2_score(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
```

```python
print(r2)
print(mse)
```

```python
import joblib
```

```python
joblib.dump(model,"zipmodel.pkl")
```
